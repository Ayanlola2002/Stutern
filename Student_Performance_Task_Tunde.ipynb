3.0 Univariate Analysis 

Univariate analysis was done to explore each of the features. It is the simplest form of analyzing data where we examine each variable individually. Univariate analysis was used to discover the kind of data (categorical or numerical),statistical description of numerical varriables with measures such as mean,std,mode,median,varriance. 

For categorical features we used  the following python codes snippet   which  calculate the number of each category in a particular variable. 


```
print('Number of gender: ', df['gender'].nunique()) 
count  = 0 
for i in df['gender'].unique(): 
    count = count + 1 
    print(count,'. ',i) 

print('Number of different races/ethnicity of people: ', df['race/ethnicity'].nunique())     
count = 0 
for i in sorted(df['race/ethnicity'].unique()): 
    count = count + 1 
    print(count, '. ',i) 

print('Number of parental Level of education: ', df['parental level of education'].nunique()) 

count = 0 
for i in df['parental level of education'].unique(): 
    count = count + 1 
    print(count, '. ', i) 
    
print('Number of lunch types: ', df['lunch'].nunique())     

count  = 0 
for i in df['lunch'].unique(): 
    count = count + 1 
    print(count,'. ',i) 

print('Number of test preparation course: ', df['test preparation course'].nunique())    
count  = 0 
for i in df['test preparation course'].unique(): 
    count = count + 1 
    print(count,'.',i) 
```

4.0 Multivariate Analysis 

Multivariate statistical analysis refers to multiple advanced techniques for examining relationships among multiple variables at the same time. In this report we use multivariate procedures in our studies to check the validity of our hypothesis.Fig 1,fig2,fig3 plots was used to validate the positivity of our hypothesis stated in the introduction section. 

4.1 Hypothesis1:How Test Preparation per gender affects The Three Scores of Student *


```
plt.figure(figsize=(20,8)) 
plt.subplot(1, 3, 1) 
sns.barplot(x='test preparation course',y='math score',data=df,hue='gender',palette='summer') 
plt.title('MATH SCORES') 
plt.subplot(1, 3, 2) 
sns.barplot(x='test preparation course',y='reading score',data=df,hue='gender',palette='summer') 
plt.title('READING SCORES') 
plt.subplot(1, 3, 3) 
sns.barplot(x='test preparation course',y='writing score',data=df,hue='gender',palette='summer') 
plt.title('WRITING SCORES') 
plt.show() 

```

Fig 3 Plots Showing test preparation per gender affects The Three Scores

From the first plot of Fig 3 we can see the math scores of boys are better irrespective of whether they completed the course or not. 
while  the next two plots shows that girls perform more better in reading and writing. From all three plots of Fig 3 its can be infered that if the course is completed the student can achieve higher scores 

4.2 Hypothesis 2:How Parental education level affects the Three Scores of Student 



```
fig,ax=plt.subplots() 
sns.barplot(x=df['parental level of education'],y='Total_Marks',data=df,palette='Wistia') 
fig.autofmt_xdate() 
```

Fig 4  Plots showing How Parental level of education Affects Scores

4.3 Hypothesis 3: How Race/Ethnicity affects the three Scores of Student 


```
plt.figure(figsize=(14,4)) 
 
plt.subplot(1,3,1) 
sns.barplot(x = 'race/ethnicity', y = 'reading score', data = df) 
plt.xticks(rotation = 90) 
 
plt.subplot(1,3,2) 
sns.barplot(x = 'race/ethnicity', y = 'writing score', data = df) 
plt.xticks(rotation = 90) 
 
plt.subplot(1,3,3) 
sns.barplot(x = 'race/ethnicity', y = 'math score', data = df) 
plt.xticks(rotation = 90) 
 
plt.tight_layout() 
```

Fig 5 Plots Showing How Race /Ethniticity of Student Affect Student Performance 
From the above plots we can infer that People from ethnic group E have higher score in all three categories, where as people from group A have the lowest score in all three categories .

4.4 Hypothesis 4:How Lunch types affects the three Scores of Student 



```
plt.figure(figsize=(14,4)) 
 
plt.subplot(1,3,1) 
sns.barplot(x = 'lunch', y = 'reading score', data = df) 
 
plt.subplot(1,3,2) 
sns.barplot(x = 'lunch', y = 'writing score', data = df) 
 
plt.subplot(1,3,3) 
sns.barplot(x = 'lunch', y = 'math score', data = df) 
 
 
plt.tight_layout() 
```

From the above plots Students who availed standard lunch have scored higher in all the three categories compared to students who have taken free/ reduced lunch.
